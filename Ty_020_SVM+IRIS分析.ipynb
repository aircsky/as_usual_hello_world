```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
```

```python
y = iris.target
```

```python
X = x[:, 2:]
```

```python
plt.scatter(X[:,0] , X[:,1] , c = y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,                                         
                                                       test_size=0.2,
                                                       random_state=30)
```

```python
plt.scatter(X_train[:,0] , X_train[:,1] , c = y_train)
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(X_train,y_train)
```

```python
y_predick = clf.predict(X_test)
```

```python
plt.scatter(X_test[:,0] , X_test[:,1] , c = y_predick)
```

```python
plt.scatter(X_test[:,0] , X_test[:,1] , c = y_predick - y_test)
```
