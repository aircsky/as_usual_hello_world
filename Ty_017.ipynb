```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

一如往常--------------

```python
x = np.linspace(0,10,100)
```

```python
y = 1.5*x+0.8+2*np.random.randn(100)
```

```python
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
線性回歸 = LinearRegression()
```

```python
X = x.reshape(100,1)
```

```python
線性回歸.fit(X,y)
```

```python
Y = 線性回歸.predict(X)
```

```python
plt.scatter(x,y)
plt.plot(x,1.5*x+0.8,"g")
plt.plot(x,Y,"r")
```
