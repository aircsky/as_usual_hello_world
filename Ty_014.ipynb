```python
df["總級分"] = df[["國文","英文","數學","自然","社會"]].sum(1)
```

sum() -> 不知道是哪方向加總

sum(0) 列加總(100人) 不是我們要的 sum(1) 行加總

```python
df
```

```python
df["加權分"] = df.國文*0.8 + df.數學*1.5 + df.英文*1.3 + df.社會*0.8 + df.自然*1.3
```

```python
df
```

```python
df.sort_values(by = "加權分" , ascending=False)
```

ascending=False = 由大到小

```python
df.sort_values(by = ["加權分" , "總級分"] , ascending=False)
```

```python
df
```

仍然沒有更改，pd請向創新資料 以免影響救有資料

```python
df.sort_values(by = ["加權分" , "總級分"] , ascending=False , inplace=True)
```
